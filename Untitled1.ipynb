```python
import yfinance as yf
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

stocks = ["CPALL.BK", "MTC.BK", "BH.BK", "ADVANC.BK", "KTC.BK", "ICHI.BK", "SISB.BK"]

def get_stock_data(tickers):
    data_list = []
    for ticker in tickers:
        print(f"กำลังดึง {ticker}...", end=" ")
        try:
            info = yf.Ticker(ticker).info
            roe = info.get("returnOnEquity")
            roa = info.get("returnOnAssets")
            de  = info.get("debtToEquity")
            div = info.get("dividendYield")
            pe  = info.get("trailingPE")
            data_list.append({
                "Symbol":        ticker.replace(".BK", ""),
                "ROE (%)":       round(roe * 100, 2) if roe else None,
                "ROA (%)":       round(roa * 100, 2) if roa else None,
                "D/E Ratio":     round(de  / 100, 2) if de  else None,
                "Div Yield (%)": round(div * 100, 2) if div else None,
                "P/E Ratio":     round(pe,        2) if pe  else None,
            })
            print("✅")
        except Exception as e:
            print(f"❌ {e}")
    return pd.DataFrame(data_list)

df = get_stock_data(stocks)
display(df.sort_values("ROE (%)", ascending=False).reset_index(drop=True))
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

colors = ['#3266ad','#17a84e','#e8a21b','#e0402a','#7b5ea7','#2ec4b6','#e76f51']

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("ROE (%)", "ROA (%)", "D/E Ratio", "Div Yield (%)"),
    vertical_spacing=0.18, horizontal_spacing=0.1
)

for (r, c), metric in zip([(1,1),(1,2),(2,1),(2,2)], ["ROE (%)","ROA (%)","D/E Ratio","Div Yield (%)"]):
    sub = df[["Symbol", metric]].dropna().sort_values(metric, ascending=False)
    fig.add_trace(go.Bar(
        x=sub["Symbol"], y=sub[metric],
        marker_color=colors[:len(sub)],
        text=sub[metric].round(2), textposition="outside",
        showlegend=False
    ), row=r, col=c)

fig.update_layout(
    height=600,
    title_text="7 จตุรเทพ — Fundamental Overview",
    paper_bgcolor="white", plot_bgcolor="#f8f9fa",
    font=dict(family="Arial", size=12), margin=dict(t=80, b=40)
)
fig.show()
```
